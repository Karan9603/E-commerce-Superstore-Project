```python
null_summary = store_data.isnull().sum()
print (null_summary)
```

```python
# Remove rows with any null values. ## in this case not needed as there is no null values

store_data_cleaned = store_data.dropna()

# Display the cleaned DataFrame

store_data_cleaned.info()

#If you want to modify the original DataFrame (instead of creating a new variable), you can use the inplace=True argument:

# Remove rows with any null values in place
# store_data.dropna(inplace=True)
```

```python
# Concatenate address columns into a single 'Full Address' column

store_data['Full Address'] = store_data['Country'] + ' ' + store_data['City'].fillna('') + ' ' + store_data['State'] + ' ' + store_data['Postal Code'].fillna('').astype(str) + ' ' + store_data['Region']

# Find customers with multiple addresses
customers_with_multiple_addresses = store_data.groupby('Customer Name')['Full Address'].nunique()
customers_with_multiple_addresses = customers_with_multiple_addresses[customers_with_multiple_addresses > 1]
print("Customers with multiple addresses:")
print(customers_with_multiple_addresses)

# Find addresses with multiple customers
addresses_with_multiple_customers = store_data.groupby('Full Address')['Customer Name'].nunique()
addresses_with_multiple_customers = addresses_with_multiple_customers[addresses_with_multiple_customers > 1]
print("\nAddresses with multiple customers:")
print(addresses_with_multiple_customers)
```

```python
# above code explaination

# The code you've provided is used to find customers who have multiple addresses in the store_data DataFrame. Let me break down the code step by step:

## Code Breakdown:

# # Find customers with multiple addresses
# customers_with_multiple_addresses = store_data.groupby('Customer Name')['Full Address'].nunique()
# store_data.groupby('Customer Name'):

# This part groups the data by the Customer Name column. It means that for each unique customer name, all rows corresponding to that customer will be grouped together.
# ['Full Address']:

# After grouping by Customer Name, we are selecting the Full Address column. This means that we are going to analyze the addresses associated with each customer.
# .nunique():

# The nunique() function counts the number of unique values in the Full Address column for each Customer Name.
# This will give us a count of how many distinct addresses each customer has in the dataset.
# For example, if a customer has two different addresses, the result for that customer will be 2. If the customer has one address, the result will be 1.
# After this line, customers_with_multiple_addresses will be a pandas Series with Customer Name as the index, and the number of unique addresses (Full Address) for each customer as the value.

# Next line:

# customers_with_multiple_addresses = customers_with_multiple_addresses[customers_with_multiple_addresses > 1]
# Here, we are filtering the Series customers_with_multiple_addresses to only include customers who have more than one unique address.
# The condition customers_with_multiple_addresses > 1 returns a Boolean Series where:
# True means the customer has more than one address.
# False means the customer has only one address.
# This line effectively filters out customers who have just one address and keeps only those who have multiple addresses.

# Finally:

# print("Customers with multiple addresses:")
# print(customers_with_multiple_addresses)
# This simply prints the result to the console. It shows the customer names who have multiple addresses along with the count of their unique addresses.
```

```python
# View detailed data for customers with multiple addresses
filtered_customers = store_data[store_data['Customer Name'].isin(customers_with_multiple_addresses.index)]
print("\nFiltered data for customers with multiple addresses:")
print(filtered_customers)

# View detailed data for addresses with multiple customers
filtered_addresses = store_data[store_data['Full Address'].isin(addresses_with_multiple_customers.index)]
print("\nFiltered data for addresses with multiple customers:")
print(filtered_addresses)
```
