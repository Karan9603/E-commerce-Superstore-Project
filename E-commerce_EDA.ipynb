```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
F_Data = pd.read_csv(r'C:\Data Analysis - Files\Project 5 - Superstore\Sample-Superstore.csv', sep=',', encoding='latin1', on_bad_lines='skip')
F_Data.head()
```

```python
# Sales each catogery
plt.figure(figsize=(10,5))
plt.subplot (1,2,1)
sns.barplot(data=F_Data , x= F_Data['Category'], y= F_Data['Sales'])
plt.title('Sales Per Catagory')
plt.subplot (1,2,2)
sns.barplot(data=F_Data , x= F_Data['Category'], y= F_Data['Profit'], color= 'green')
plt.title('Profit Per Catagory')
```

```python
#What customer purchasing patterns can you discover? 
#Such as activity during the days of the week, weekly, monthly, quarterly, yearly, etc.

F_Data.info()
F_Data['Order Date'] = pd.to_datetime(F_Data['Order Date'])
F_Data['Weekdays'] = pd.to_datetime(F_Data['Order Date']).dt.day_of_week
day = {0 : 'Monday',
       1 : 'Tuesday',
       2 : 'Wednesday',
       3 : 'Thursday',
       4 : 'Friday',
       5 : 'Saturday',       
       6 : 'Sunday'  }
F_Data['Weekdays']=F_Data['Weekdays'].map(day)
order = ('Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday')
F_Data['Weekdays']=pd.Categorical(F_Data['Weekdays'], categories=order , ordered=True)
```

```python
F_Data.head()
```

```python

sns.barplot(data=F_Data , x= F_Data['Weekdays'], y= F_Data['Sales'])
plt.title('Sales Per Weekday')
```
